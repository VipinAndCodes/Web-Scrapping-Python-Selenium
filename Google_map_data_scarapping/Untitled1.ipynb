```python
import json
import requests
```

```python
api_file = open("API_KEY.txt","r")
api_key = api_file.read()
api_file.close()
```

```python
location = input("location")
radius = input("Enter radius")
type = input("enter location type")
url ="https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

r=requests.get(url+"location="+ location + "&radius="+ radius + "&type="+ type +"&key="+api_key)
```

```python
response = r.json()
```
